## Load data

```python
path = "../../data/MRW.npz"
data = np.load(path)
print(data.files)
```

## Explore data

```python
Y_LABELS = data[sctes.Y_LABELS]
print(f"{len(Y_LABELS) = !s}")
print(f"{Y_LABELS = !s}")
```

```python
Y = data[sctes.Y]
```

```python
print(f"{Y.shape = !s}")
print(f"S.shape[-1] == len(Y_LABELS) ? : {Y.shape[-1] == len(Y_LABELS)}")
print(f"{Y.dtype = }")
```

```python
X = data[sctes.X]
```

```python
N, LEN = X.shape
```

```python
print(f"{X.shape = !s}")
print(f"X.shape[-1] == 2**15 ? : {X.shape[-1] == 2**15}")
print(f"{X.dtype = }")
```

## Visualize data

### Random samples

```python
nb_plots = 1
index = np.random.choice(X.shape[0], nb_plots, replace=False)  
X_plots = X[index,:]
Y_plots = Y[index, :]
for i in range(nb_plots):
    x = np.linspace(1,LEN,LEN)
    y = X_plots[i,:]
    c1, c2, L, epsilon = Y_plots[i]
    plt.figure(figsize = (18, 5))
    plt.subplot(nb_plots, 1, (i+1))
    plt.plot(x, y)
    c1_rounded = round(c1, 2)
    c2_rounded = round(c2, 2)
    plt.title(f"{c1_rounded = !s}, {c2_rounded = !s}, {L = !s}, {epsilon = !s}")
    plt.grid()
plt.show()
```

### Plot the 32 samples with same statistics

```python
n_subsamples = 32
X_plots = X[:n_subsamples,:]
Y_plots = Y[:n_subsamples, :]
for i in range(n_subsamples):
    x = np.linspace(1,LEN,LEN)
    y = X_plots[i,:]
    c1, c2, L, epsilon = Y_plots[i]
    plt.figure(figsize=((20,100)))
    plt.subplot(n_subsamples, 1, (i+1))
    plt.plot(x, y)
    plt.title(f"{c1 = !s}, {c2 = !s}, {L = !s}, {epsilon = !s}")
    plt.grid()
plt.show()
```
