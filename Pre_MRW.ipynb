# Data_MRW.npz : Exploratory Analysis

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,10)
```

## Load data

```python
data = np.load("Pre_MRW.npz")
print(data.files)
```

## Explore data

```python
c1s = data["c1s"]
print(f"{len(c1s) = !s}")
print(f"{c1s = !s}")
```

```python
c2s = data["c2s"]
print(f"{len(c2s) = !s}")
print(f"{c2s = !s}")
```

```python
Ls = data["Ls"]
print(f"{len(Ls) = !s}")
print(f"{Ls = !s}")
```

```python
epsilons = data["epsilons"]
print(f"{len(epsilons) = !s}")
print(f"{epsilons = !s}")
```

```python
N = data["N"]
print(f"{N = !s}")
```

```python
win = data["win"]
print(f"{win = !s}")
```

```python
mrw = data["mrw"]
```

```python
print(f"{mrw.shape = !s}")
print(f"mrw.shape[0] == len(c1s)      ? : {mrw.shape[0] == len(c1s)}")
print(f"mrw.shape[1] == len(c2s)      ? : {mrw.shape[1] == len(c2s)}")
print(f"mrw.shape[2] == len(Ls)       ? : {mrw.shape[2] == len(Ls)}")
print(f"mrw.shape[3] == len(epsilons) ? : {mrw.shape[3] == len(epsilons)}")
print(f"mrw.shape[4] == N             ? : {mrw.shape[4] == N}")
```

## Visualize data

```python
x = np.linspace(1,N,N)
```

```python
for ic1 in range(len(c1s)):
    plt.figure()
    plt.subplot(len(c1s), 1, ic1+1)
    ic1 = ic1
    ic2 = 0
    iL = 0
    iepsilon = 0
    plt.plot(x, mrw[ic1, ic2, iL, iepsilon, :])
    plt.title(f"C1: {c1s[ic1]}, C2: {c2s[ic2]}, L: {Ls[iL]}, Epsilon: {epsilons[iepsilon]}")
    plt.grid()
plt.show()
```

```python
for ic2 in range(len(c2s)):
    plt.figure()
    plt.subplot(len(c2s), 1, ic2+1)
    ic1 = 0
    ic2 = ic2
    iL = 0
    iepsilon = 0
    plt.plot(x, mrw[ic1, ic2, iL, iepsilon, :])
    plt.title(f"C1: {c1s[ic1]}, C2: {c2s[ic2]}, L: {Ls[iL]}, Epsilon: {epsilons[iepsilon]}")
    plt.grid()
plt.show()
```

```python
for iL in range(len(Ls)):
    plt.figure()
    plt.subplot(len(Ls), 1, iL+1)
    ic1 = 0
    ic2 = 0
    iL = iL
    iepsilon = 0
    plt.plot(x, mrw[ic1, ic2, iL, iepsilon, :])
    plt.title(f"C1: {c1s[ic1]}, C2: {c2s[ic2]}, L: {Ls[iL]}, Epsilon: {epsilons[iepsilon]}")
    plt.grid()
plt.show()
```

```python
for iepsilon in range(len(epsilons)):
    plt.figure()
    plt.subplot(len(epsilons), 1, iepsilon+1)
    ic1 = 0
    ic2 = 0
    iL = 0
    iepsilon = iepsilon
    plt.plot(x, mrw[ic1, ic2, iL, iepsilon, :])
    plt.title(f"C1: {c1s[ic1]}, C2: {c2s[ic2]}, L: {Ls[iL]}, Epsilon: {epsilons[iepsilon]}")
    plt.grid()
plt.show()
```
